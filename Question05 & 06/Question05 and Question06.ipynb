```python
import pandas as pd
#Create a new data-frame for classifying some countries into some categories
data = {'Country_Name': ['United States', 'China', 'India', 'Hong Kong', 'Taiwan', 'Canada', 'Brazil', 'Japan', 'North Korea', 'United Kingdom', 'Sweden', 'Turkey', 'Ukraine', 'Poland' ],
        'Continent': ['North America', 'Asia', 'Asia', 'Asia', 'Asia', 'North America', 'South America', 'Asia', 'Asia', 'Europe','Europe','Europe','Europe','Europe']
       }
df = pd.DataFrame(data)
df
```

```python
dvd_df = pd.read_csv('Query05_result.csv')
#Left Merge to join the last dataframe with dvd db
merged_df = pd.merge(df, dvd_df, left_on='Country_Name', right_on='country', how='left') 
merged_df.drop(columns=['Country_Name'],inplace=True)
sorted_df = merged_df.sort_values(by='Continent')
sorted_df
```
