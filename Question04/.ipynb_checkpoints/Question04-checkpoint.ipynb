```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Query04_result.csv')
# Set the figure size
plt.figure(figsize=(10, 5))

plt.bar(df['full_name'], df['total_revenue'])
#To give the title of the graph
plt.title('Top 10 Actors Who Are Bringing In The Most Revenue', fontsize=15, weight='bold')

#To customize the graph
plt.xlabel('Actors',fontsize=12, weight='bold')
plt.ylabel('Total Revenue ($)', fontsize=12, weight='bold')
plt.xticks(rotation=30)

plt.show()
```
