```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('Query01_result.csv')
df['date'] = pd.to_datetime(df['date'])

# Dataframe for three different countries
df_china = (df.where((df['country'] == 'China'))).dropna()
df_us = (df.where((df['country'] == 'United States'))).dropna()
df_india = (df.where((df['country'] == 'India'))).dropna()

# Set the figure size
plt.figure(figsize=(18, 6))

#To give the name of the chart
plt.title('Daily Revenue For Three Countries', fontsize=15, weight='bold')


# Adjust the number of x-axis ticks
plt.locator_params(axis='x', nbins=4)

plt.plot_date(df_china['date'], df_china['revenue_per_day'], color='Red', label='China')
plt.plot_date(df_us['date'], df_us['revenue_per_day'], color='Blue', label='United States')
plt.plot_date(df_india['date'], df_india['revenue_per_day'], color='Green', label='India')

#To give the name of the x-axis and y-axis
plt.xlabel('Date',fontsize=12, weight='bold')
plt.ylabel('Daily Revenue ($)',fontsize=12, weight='bold')
plt.legend()

plt.show()
```
