```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Query03_result.csv')
df_sorted = df.sort_values(by='total_revenue')
# Set the figure size
plt.figure(figsize=(14, 28))
#Plotting the bar graph
plt.barh(df_sorted['country'], df_sorted['total_revenue'])

#Customize the plot
plt.xlabel('Total Revenue ($)', fontsize = 15, weight='bold')
plt.ylabel('Countries', fontsize = 15, weight='bold')
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.title('Total Revenue From Different Countries To The DVD Store', fontsize = 20, weight='bold')
plt.tight_layout()

plt.show()
```
