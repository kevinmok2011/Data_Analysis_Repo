```python
import pandas as pd

df = pd.read_csv('Query02_result.csv')

def sentiment_detection(sentence):
    positive = False
    negative = False
    for i in ('Amazing', 'Insightful', 'Touching', 'Fateful', 'Epic','Brilliant', 'Unbelieveable', 'Astounding', 'Awe-inspiring', 'Stunning'):
        if sentence.count(i) > 0:
            positive = True
    for i in ('Boring', 'Thrilling', 'Lacklustre', 'Fanciful'):
        if sentence.count(i) > 0:
            negative = True
        if positive == negative == False:
            return 'neutral'
        elif positive == True != negative:
            return 'positive'
        elif negative == True != positive:
            return 'negative'
        else:
            return 'mixed'
```

```python
df['sentiment_detection'] = df['description'].apply(sentiment_detection)
df
```
